```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = {
    'size': [500, 1000, 1500, 2000, 2500],
    'price': [25, 50, 75, 100, 125]
}
df = pd.DataFrame(data)
```

```python
model = LinearRegression()
model.fit(df[['size']], df['price'])
```

```python
size_input = 1800
predicted_price = model.predict([[size_input]])
print(f"Predicted price for {size_input} sq.ft = â‚¹{predicted_price[0]} lakhs")
```

```python
plt.scatter(df['size'], df['price'], color='blue', label='Training data')
plt.plot(df['size'], model.predict(df[['size']]), color='red', label='Prediction Line')
plt.xlabel('Size (sq ft)')
plt.ylabel('Price (lakhs)')
plt.title('House Price Prediction')
plt.legend()
plt.show()
```
